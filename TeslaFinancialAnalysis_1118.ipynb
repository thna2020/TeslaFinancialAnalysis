## Profitability Analysis

```python
profit_df = pd.DataFrame(columns = ['Year', 'Gross Profit Percentage (%)', 'Return on Sales (%)', 
                                   'Asset Turnover ($)', 'Return on Assets (%)',
                                   'Return on Common Stockholders\' Equity (%)'])
```

```python
# year
profit_df['Year'] = cf_df['Year'].values

# gross profit percentage (gpp)
gpp_1013 = is_df['Gross profit'] / is_df['Total revenues']
gpp_1418 = is_df2['Gross profit'] / is_df2['Total revenues']
profit_df['Gross Profit Percentage (%)'] = gpp_1013.append(gpp_1418).values

# return on sales (ros)
ros_1013 = is_df['Net income'] / is_df['Total revenues']
ros_1418 = is_df2['Net income'] / is_df2['Total revenues']
profit_df['Return on Sales (%)'] = ros_1013.append(ros_1418).values

# asset turnover (ato)
avg_assets = bs_df['Total assets'].rolling(window=2).mean()
ato_1013 = is_df['Total revenues'] / avg_assets[:4].values
ato_1418 = is_df2['Total revenues'] / avg_assets[4:].values
profit_df['Asset Turnover ($)'] = ato_1013.append(ato_1418).values

# return on assets (roa)
roa_1013 = is_df['Net income'] / avg_assets[:4].values
roa_1418 = is_df2['Net income'] / avg_assets[4:].values
profit_df['Return on Assets (%)'] = roa_1013.append(roa_1418).values

# return on common stockholders' equity (roe)
avg_equity = bs_df['Total stockholder\'s equity'].rolling(window=2).mean()
roe_1013 = is_df['Net income'] / avg_equity[:4].values
roe_1418 = is_df2['Net income'] / avg_equity[4:].values
profit_df['Return on Common Stockholders\' Equity (%)'] = roe_1013.append(roe_1418).values

profit_df
```

```python
# profitability analysis from 2011 to 2018
profit_df2 = profit_df[1:]
profit_df2.reset_index(drop='True')
```

```python
plt.bar(profit_df2['Year'], profit_df2['Return on Sales (%)'], color='lightcoral')
plt.bar(profit_df2['Year'], profit_df2['Gross Profit Percentage (%)'], color='forestgreen')
plt.show()
```

```python
plt.bar(profit_df2['Year'], profit_df2['Asset Turnover ($)'], color='forestgreen')
plt.bar(profit_df2['Year'], profit_df2['Return on Assets (%)'], color='lightcoral')
plt.show()
```

```python
plt.plot(profit_df2['Year'], profit_df2['Return on Common Stockholders\' Equity (%)'])
plt.show()
```
